```python
"""
MH-FedXAI: Mental Health - Federated Explainable AI
5 Local Clients | SMOTE | Binary Classification
Targets: Depression (Yes/No), Anxiety (Yes/No), Stress (Yes/No)

Metrics: Accuracy, Precision, Recall, F1, MCC, P-Value, ROC-AUC
Visuals: Confusion Matrix (binary), ROC Curve
XAI: SHAP (Global Importance), LIME (Local Interpretation)
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from scipy.stats import chi2_contingency
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    matthews_corrcoef, confusion_matrix, roc_curve, auc,
    roc_auc_score, classification_report, ConfusionMatrixDisplay
)
from sklearn.ensemble import (
    RandomForestClassifier, AdaBoostClassifier, StackingClassifier
)
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from imblearn.over_sampling import SMOTE

# --- XAI Imports ---
import shap
import lime
import lime.lime_tabular

warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid", context="paper", font_scale=1.1)
plt.rcParams['font.family'] = 'serif'

# Ensure all pandas float output is 4 decimal places
pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:,.4f}'.format

NUM_CLIENTS = 5

# ==================== UTILITY FUNCTIONS ====================

def print_header(title):
    print(f"\n{'='*80}\n{title.center(80)}\n{'='*80}")


def calculate_p_value(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    try:
        chi2, p_value, dof, expected = chi2_contingency(cm)
        return p_value
    except:
        return np.nan


def calculate_all_metrics(y_true, y_pred, y_prob=None, model_name="Model"):
    metrics = {
        'Model': model_name,
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred, average='weighted', zero_division=0),
        'Recall': recall_score(y_true, y_pred, average='weighted', zero_division=0),
        'F1 Score': f1_score(y_true, y_pred, average='weighted', zero_division=0),
        'MCC': matthews_corrcoef(y_true, y_pred),
        'P-Value': calculate_p_value(y_true, y_pred)
    }
    if y_prob is not None and len(np.unique(y_true)) == 2:
        try:
            metrics['ROC-AUC'] = roc_auc_score(y_true, y_prob[:, 1])
        except:
            metrics['ROC-AUC'] = np.nan
    return metrics


# ==================== DATA LOADING & PREPROCESSING ====================

print_header("PHASE 1: DATA LOADING & PREPROCESSING")

try:
    data = pd.read_csv('/kaggle/input/dass-42-dataset/DASS-42.csv',
                       on_bad_lines='skip', delimiter='\t')
    print("✓ Dataset loaded successfully")
except Exception as e:
    print(f"⚠ Error loading dataset: {e}")
    raise

data = data.dropna().drop_duplicates()
print(f"✓ Data cleaned: {data.shape[0]} samples, {data.shape[1]} features")

depression_questions = ['Q3A', 'Q5A', 'Q10A', 'Q13A', 'Q16A', 'Q17A', 'Q21A',
                        'Q24A', 'Q26A', 'Q31A', 'Q34A', 'Q37A', 'Q38A', 'Q42A']
anxiety_questions    = ['Q2A', 'Q4A', 'Q7A', 'Q9A', 'Q15A', 'Q19A', 'Q20A',
                        'Q23A', 'Q25A', 'Q28A', 'Q30A', 'Q36A', 'Q40A', 'Q41A']
stress_questions     = ['Q1A', 'Q6A', 'Q8A', 'Q11A', 'Q12A', 'Q14A', 'Q18A',
                        'Q22A', 'Q27A', 'Q29A', 'Q32A', 'Q33A', 'Q35A', 'Q39A']
additional_features  = ['age', 'education', 'religion', 'married']

for questions in [depression_questions, anxiety_questions, stress_questions]:
    data[questions] = data[questions].apply(pd.to_numeric, errors='coerce').fillna(0)

for feat in additional_features:
    if feat in data.columns:
        if data[feat].dtype == 'object':
            data[feat] = LabelEncoder().fit_transform(data[feat].astype(str))
        else:
            data[feat] = data[feat].fillna(data[feat].median())

# ==================== BINARY TARGET CREATION ====================

def make_binary_target(scores, threshold, name):
    target = (scores > threshold).astype(int)
    if len(np.unique(target)) < 2:
        median_thresh = scores.median()
        print(f"  ⚠ WARNING: threshold={threshold} gave only 1 class for {name}. "
              f"Falling back to median threshold={median_thresh:.1f}")
        target = (scores > median_thresh).astype(int)
    if len(np.unique(target)) < 2:
        mean_thresh = scores.mean()
        print(f"  ⚠ WARNING: median fallback still gave 1 class for {name}. "
              f"Using mean threshold={mean_thresh:.1f}")
        target = (scores > mean_thresh).astype(int)
    return target

data['Depression_score'] = data[depression_questions].sum(axis=1)
data['Anxiety_score']    = data[anxiety_questions].sum(axis=1)
data['Stress_score']     = data[stress_questions].sum(axis=1)

data['Depression_target'] = make_binary_target(data['Depression_score'],  9, 'Depression')
data['Anxiety_target']    = make_binary_target(data['Anxiety_score'],     7, 'Anxiety')
data['Stress_target']     = make_binary_target(data['Stress_score'],     14, 'Stress')

label_names = {
    'Depression': ['No', 'Yes'],
    'Anxiety':    ['No', 'Yes'],
    'Stress':     ['No', 'Yes']
}

print("✓ Binary target variables created:")
for t in ['Depression', 'Anxiety', 'Stress']:
    vc = data[f'{t}_target'].value_counts()
    print(f"  {t}: No={vc.get(0, 0)}, Yes={vc.get(1, 0)}")

# ==================== MODEL DEFINITION ====================

def create_mh_fedxai_model():
    base_learners = [
        ('rf',  RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)),
        ('svm', SVC(kernel='rbf', probability=True, C=1.0, random_state=42)),
        ('ada', AdaBoostClassifier(n_estimators=100, learning_rate=1.0, random_state=42)),
        ('lr',  LogisticRegression(max_iter=1000, random_state=42)),
        ('knn', KNeighborsClassifier(n_neighbors=5)),
        ('gnb', GaussianNB())
    ]
    meta_learner = LogisticRegression(max_iter=1000, random_state=42)
    return StackingClassifier(
        estimators=base_learners,
        final_estimator=meta_learner,
        cv=5,
        n_jobs=-1
    )

# ==================== FEDERATED LEARNING HELPERS ====================

def split_data_into_clients(X, y, num_clients=5, random_state=42):
    np.random.seed(random_state)
    indices = np.random.permutation(len(X))
    shards = np.array_split(indices, num_clients)
    clients = []
    for shard in shards:
        clients.append((X.iloc[shard].reset_index(drop=True), y[shard]))
    return clients


def federated_average(model_weights_list):
    coefs = np.array([w['coef'] for w in model_weights_list])
    intercepts = np.array([w['intercept'] for w in model_weights_list])
    return {
        'coef': coefs.mean(axis=0),
        'intercept': intercepts.mean(axis=0)
    }


# ==================== VISUALIZATION FUNCTIONS ====================

def plot_confusion_matrix_binary(y_true, y_pred, target_name, class_names, title=""):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    disp.plot(cmap='Blues', values_format='d', ax=plt.gca())
    plt.title(f"Confusion Matrix — {target_name} {title}", fontsize=13, fontweight='bold')
    plt.tight_layout()
    plt.show()


def plot_roc_curve(y_true, y_prob, target_name, title=""):
    fpr, tpr, _ = roc_curve(y_true, y_prob[:, 1])
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(7, 5))
    plt.plot(fpr, tpr, color='darkorange', lw=2,
             label=f'ROC curve (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--',
             label='Random Classifier')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f"ROC Curve — {target_name} {title}", fontsize=13, fontweight='bold')
    plt.legend(loc="lower right")
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()


# ==================== MAIN FEDERATED LOOP ====================

print_header("PHASE 2: FEDERATED TRAINING — 5 LOCAL CLIENTS")

all_global_results = {}
all_client_results = {}

for target_name, questions in [
    ('Depression', depression_questions),
    ('Anxiety',    anxiety_questions),
    ('Stress',     stress_questions)
]:
    print(f"\n{'='*80}")
    print(f"TARGET: {target_name.upper()}")
    print(f"{'='*80}")

    features = questions + [f for f in additional_features if f in data.columns]
    X_all = data[features].copy().astype(float)
    y_all = data[f'{target_name}_target'].copy().values

    # Safety check before SMOTE
    unique, counts = np.unique(y_all, return_counts=True)
    if len(unique) < 2:
        print(f"  ⚠ Skipping {target_name} — only 1 class present.")
        continue

    min_class_count = counts.min()
    k_neighbors = min(5, min_class_count - 1)
    if k_neighbors < 1:
        print(f"  ⚠ Skipping {target_name} — minority class too small for SMOTE.")
        continue

    # ---- SMOTE on full dataset, then split into clients ----
    smote = SMOTE(random_state=42, k_neighbors=k_neighbors)
    X_res, y_res = smote.fit_resample(X_all, y_all)
    X_res_df = pd.DataFrame(X_res, columns=features)

    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X_res_df), columns=features)

    print(f"✓ After SMOTE: {X_scaled.shape[0]} samples | Classes: {np.bincount(y_res)}")

    # Global test set (held out before client split)
    X_train_global, X_test_global, y_train_global, y_test_global = train_test_split(
        X_scaled, y_res, test_size=0.2, random_state=42, stratify=y_res
    )

    client_data = split_data_into_clients(X_train_global, y_train_global, num_clients=NUM_CLIENTS)

    client_metrics_list = []
    model_weights_list  = []
    client_models       = []

    # ---------- LOCAL CLIENT TRAINING ----------
    print(f"\n{'─'*60}")
    print(f"  LOCAL CLIENT TRAINING")
    print(f"{'─'*60}")

    for client_id, (X_client, y_client) in enumerate(client_data):
        if len(np.unique(y_client)) < 2:
            print(f"  [Client {client_id + 1}] ⚠ Skipped — only 1 class.")
            continue

        print(f"\n  [Client {client_id + 1}] Samples: {len(X_client)} | "
              f"Class dist: {np.bincount(y_client)}")

        X_c_train, X_c_test, y_c_train, y_c_test = train_test_split(
            X_client, y_client, test_size=0.2, random_state=42, stratify=y_client
        )

        local_model = create_mh_fedxai_model()
        local_model.fit(X_c_train, y_c_train)

        y_c_pred = local_model.predict(X_c_test)
        y_c_prob = local_model.predict_proba(X_c_test)

        m = calculate_all_metrics(y_c_test, y_c_pred, y_c_prob, model_name=f"Client {client_id + 1}")
        client_metrics_list.append(m)
        client_models.append((client_id, local_model, X_client, y_client))

        auc_str = f"{m.get('ROC-AUC', np.nan):.4f}" if not np.isnan(m.get('ROC-AUC', np.nan)) else "N/A"
        print(f"    Accuracy:  {m['Accuracy']:.4f} | F1: {m['F1 Score']:.4f} | "
              f"MCC: {m['MCC']:.4f} | AUC: {auc_str}")

        meta = local_model.final_estimator_
        model_weights_list.append({'coef': meta.coef_.copy(), 'intercept': meta.intercept_.copy()})

    if not client_models:
        print(f"  ⚠ No valid client models for {target_name}. Skipping.")
        continue

    all_client_results[target_name] = client_metrics_list

    # ---------- INDIVIDUAL CLIENT RESULTS TABLE ----------
    print(f"\n{'─'*80}")
    print(f"  INDIVIDUAL LOCAL CLIENT MODEL PERFORMANCE — {target_name.upper()}")
    print(f"{'─'*80}")
    client_df = pd.DataFrame(client_metrics_list).set_index('Model')
    display_cols = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'MCC', 'P-Value']
    if 'ROC-AUC' in client_df.columns:
        display_cols.append('ROC-AUC')
    print(client_df[display_cols].to_string())

    # ---------- FEDERATED AGGREGATION ----------
    avg_weights = federated_average(model_weights_list)
    global_model = client_models[0][1]
    global_model.final_estimator_.coef_      = avg_weights['coef']
    global_model.final_estimator_.intercept_ = avg_weights['intercept']

    # ---------- GLOBAL MODEL EVALUATION ----------
    y_pred_global = global_model.predict(X_test_global)
    y_prob_global = global_model.predict_proba(X_test_global)

    global_metrics = calculate_all_metrics(
        y_test_global, y_pred_global, y_prob_global,
        model_name=f"MH-FedXAI Global ({target_name})"
    )
    all_global_results[target_name] = global_metrics

    print(f"\n{'─'*80}")
    print(f"  GLOBAL FEDERATED MODEL — {target_name.upper()}")
    print(f"{'─'*80}")
    print(f"  Accuracy:   {global_metrics['Accuracy']:.4f}")
    print(f"  Precision:  {global_metrics['Precision']:.4f}")
    print(f"  Recall:     {global_metrics['Recall']:.4f}")
    print(f"  F1 Score:   {global_metrics['F1 Score']:.4f}")
    print(f"  MCC:        {global_metrics['MCC']:.4f}")
    print(f"  P-Value:    {global_metrics['P-Value']:.4f}")
    if 'ROC-AUC' in global_metrics:
        print(f"  ROC-AUC:    {global_metrics['ROC-AUC']:.4f}")

    # ==================== VISUALIZATIONS ====================
    print_header(f"PHASE 3: VISUALIZATIONS — {target_name.upper()}")

    # 1. Global Confusion Matrix
    plot_confusion_matrix_binary(y_test_global, y_pred_global, target_name,
                                 label_names[target_name], "(Global — FedAvg)")

    # 2. Client Confusion Matrices
    fig, axes = plt.subplots(1, len(client_models), figsize=(5 * len(client_models), 4))
    if len(client_models) == 1: axes = [axes]
    fig.suptitle(f"Confusion Matrix — {target_name} (Per Local Client)", fontsize=14, fontweight='bold')
    for plot_idx, (cid, model, X_client, y_client) in enumerate(client_models):
        X_c_train, X_c_test, y_c_train, y_c_test = train_test_split(
            X_client, y_client, test_size=0.2, random_state=42, stratify=y_client
        )
        y_c_pred_local = model.predict(X_c_test)
        cm = confusion_matrix(y_c_test, y_c_pred_local)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_names[target_name])
        disp.plot(cmap='Blues', values_format='d', ax=axes[plot_idx], colorbar=False)
        axes[plot_idx].set_title(f"Client {cid + 1}", fontsize=11, fontweight='bold')
    plt.tight_layout()
    plt.show()

    # 3. Global ROC
    plot_roc_curve(y_test_global, y_prob_global, target_name, "(Global — FedAvg)")

    # 4. Client ROCs
    plt.figure(figsize=(9, 6))
    colors_roc = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12', '#9b59b6']
    for plot_idx, (cid, model, X_client, y_client) in enumerate(client_models):
        X_c_train, X_c_test, y_c_train, y_c_test = train_test_split(
            X_client, y_client, test_size=0.2, random_state=42, stratify=y_client
        )
        y_c_prob_local = model.predict_proba(X_c_test)
        fpr, tpr, _ = roc_curve(y_c_test, y_c_prob_local[:, 1])
        roc_auc_val = auc(fpr, tpr)
        plt.plot(fpr, tpr, color=colors_roc[plot_idx % len(colors_roc)], lw=2,
                 label=f'Client {cid + 1} (AUC={roc_auc_val:.4f})')
    plt.plot([0, 1], [0, 1], 'k--', lw=1.5, label='Random')
    plt.xlim([0, 1]); plt.ylim([0, 1.05])
    plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
    plt.title(f"ROC Curves — {target_name} (Per Local Client)", fontsize=13, fontweight='bold')
    plt.legend(loc='lower right'); plt.grid(alpha=0.3); plt.tight_layout()
    plt.show()

    # 5. Bar Chart
    metrics_bar = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'MCC']
    client_bar_df = pd.DataFrame(client_metrics_list).set_index('Model')[metrics_bar]
    client_bar_df.plot(kind='bar', figsize=(12, 5), colormap='tab10', edgecolor='black', linewidth=0.5)
    plt.title(f"Local Client Performance — {target_name}", fontsize=14, fontweight='bold')
    plt.ylabel('Score'); plt.xticks(rotation=0); plt.ylim(0, 1.1)
    plt.grid(axis='y', alpha=0.3); plt.legend(loc='lower right')
    plt.tight_layout()
    plt.show()

    # ==================== XAI EXPLAINABILITY ====================
    print_header(f"PHASE 4: XAI INTERPRETABILITY — {target_name.upper()}")

    # 1. SHAP
    print(f"  Calculating SHAP values for Global Model ({target_name})...")
    try:
        # Using a small summary of background data for kernel explainer speed
        background_data = shap.kmeans(X_train_global, 10)
        explainer = shap.KernelExplainer(global_model.predict_proba, background_data)
        
        # Take a small slice of test data for explanation (e.g. 50 samples)
        test_slice = X_test_global.iloc[:50]
        
        # Calculate SHAP values
        shap_values = explainer.shap_values(test_slice, nsamples=100)
        
        # FIXED: Check shape and select the positive class matrix if it's a list
        if isinstance(shap_values, list):
            # Binary classification returns [class0_matrix, class1_matrix]
            # We want Class 1 (Yes)
            shap_val_to_plot = shap_values[1]
        else:
            # If shape is (samples, features, classes) or just (samples, features)
            if len(np.array(shap_values).shape) == 3:
                 shap_val_to_plot = shap_values[:, :, 1]
            else:
                 shap_val_to_plot = shap_values

        print("  ✓ SHAP Summary Plot generated.")
        plt.figure(figsize=(10, 6))
        shap.summary_plot(shap_val_to_plot, test_slice, plot_type="bar", show=False)
        plt.title(f"SHAP Feature Importance (Global) — {target_name}", fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()

        print("  ✓ SHAP Beeswarm Plot generated.")
        plt.figure(figsize=(10, 6))
        shap.summary_plot(shap_val_to_plot, test_slice, show=False)
        plt.title(f"SHAP Beeswarm (Global) — {target_name}", fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"  ⚠ Error in SHAP calculation: {e}")

    # 2. LIME
    print(f"\n  Calculating LIME explanations for Global Model ({target_name})...")
    try:
        lime_explainer = lime.lime_tabular.LimeTabularExplainer(
            training_data=np.array(X_train_global),
            feature_names=features,
            class_names=label_names[target_name],
            mode='classification'
        )

        sample_idx = 0  # Explain the first sample from test set
        sample_instance = X_test_global.iloc[sample_idx]
        
        # LIME expects numpy array for the instance
        lime_exp = lime_explainer.explain_instance(
            data_row=np.array(sample_instance),
            predict_fn=global_model.predict_proba
        )
        
        plt.figure(figsize=(10, 5))
        lime_exp.as_pyplot_figure()
        plt.title(f"LIME Local Explanation — {target_name}", fontsize=12, fontweight='bold')
        plt.tight_layout()
        plt.show()
        
    except Exception as e:
        print(f"  ⚠ Error in LIME calculation: {e}")


# ==================== SUMMARY ====================

print_header("PHASE 5: GLOBAL SUMMARY ACROSS ALL TARGETS")

if all_global_results:
    global_df = pd.DataFrame(all_global_results).T
    print("\n>>> GLOBAL MODEL PERFORMANCE (After FedAvg)")
    cols = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'MCC', 'P-Value']
    if 'ROC-AUC' in global_df.columns:
        cols.append('ROC-AUC')
    print(global_df[cols].to_string())

    summary_cols = [c for c in cols[:6] if c in global_df.columns]
    fig, axes = plt.subplots(1, len(summary_cols), figsize=(18, 5))
    fig.suptitle('MH-FedXAI Global Performance — Depression, Anxiety, Stress',
                 fontsize=14, fontweight='bold')
    colors_bar = ['#2ecc71', '#3498db', '#e74c3c']
    for idx, metric in enumerate(summary_cols):
        ax = axes[idx]
        vals = global_df[metric].values
        ax.bar(global_df.index, vals, color=colors_bar[:len(global_df)])
        ax.set_title(metric, fontsize=11, fontweight='bold')
        ax.set_ylim(0, max(vals) * 1.25 if metric == 'P-Value' else 1.1)
        for i, v in enumerate(vals):
            ax.text(i, v + 0.01, f'{v:.4f}', ha='center', fontsize=9, fontweight='bold')
        ax.grid(axis='y', alpha=0.3)
        ax.tick_params(axis='x', labelsize=9)
    plt.tight_layout()
    plt.show()
else:
    print("⚠ No global results to display.")

print("\n" + "="*80)
print("COMPLETE — MH-FedXAI | 5 Local Clients | SMOTE | XAI (SHAP & LIME)")
print("="*80)
```
